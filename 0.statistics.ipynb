```python
import os
import numpy as np
current = os.getcwd()
video_imgs_path = os.path.join(current, 'ucf_image')
folders = os.listdir(video_imgs_path)
folders = natsort.natsorted(folders)
frames = dict()
for folder in folders:
        path = os.path.join(video_imgs_path,folder)
        frames[folder] = len(os.listdir(path))
```

```python
min(list(frames.values())), max(list(frames.values()))
```

```python
import pandas as pd

df = pd.DataFrame(frames.values(), frames.keys(), columns=['Number of Frames'])
label = None
counts = {}
for name in list(frames.keys()):
    if label == None:
        loc1 = name.find('_')
        loc2 = loc1 + name[loc1+1:].find('_')
        label = name[loc1+1:loc2+1]
    counts[label] = []
    
for name in list(frames.keys()):
    if label == None:
        loc1 = name.find('_')
        loc2 = loc1 + name[loc1+1:].find('_')
        label = name[loc1+1:loc2+1]

    if label in name:
        counts[label].append(frames[name])
```
